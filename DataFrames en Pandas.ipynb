### Introduccion a DataFrames en Pandas <a id='id1'> </a>

```python
import pandas as pd
```

```python
Alumno1 = pd.Series({'Nombre':'Juan',
                     'Fecha Nacimiento':'10-05-2000',
                     'Carrera':'IINF',
                     'Edad':20})
Alumno2 = pd.Series({'Nombre':'Karla',
                     'Fecha Nacimiento':'10-02-2001',
                     'Carrera':'ISIC',
                     'Edad':19})
Alumno3 = pd.Series({'Nombre':'Pedro',
                     'Fecha Nacimiento':'18-11-1998',
                     'Carrera':'IINF',
                     'Edad':22})
Alumno4 = pd.Series({'Nombre':'Maria',
                     'Fecha Nacimiento':'15-09-1999',
                     'Carrera':'IINF',
                     'Edad':21})
```

```python
Alumno1
```

```python
Alumnos = pd.DataFrame([Alumno1,Alumno2,Alumno3,Alumno4])
Alumnos
```

```python
Alumnos.iloc[1]
```

### Logind CSV as DataFrame <a id='id5'><a/>

```python
df = pd.read_csv('olympics.csv')
df
```

```python
df.head(10)
```

```python
df = pd.read_csv('olympics.csv', index_col =0, skiprows=1)
df.head(10)
```

```python
df.columns
```

```python
for col in df.columns:
    if col[:2] =='01': 
        df.rename(columns={col:'Gold' + col[4:]}, inplace = True)
    if col[:2] =='02': 
        df.rename(columns={col:'Silver' + col[4:]}, inplace = True)
    if col[:2] =='03': 
        df.rename(columns={col:'Bronze' + col[4:]}, inplace = True)
    if col[:1] =='№': 
        df.rename(columns={col:'#' + col[1:]}, inplace = True)
```

```python
df.head(10)
```

### Busquedas en DataFrames <a id='id6'><a/>

```python
ganadores_oro = df.where(df['Gold'] > 0)
ganadores_oro.head(10)
```
