```python
import pandas as pd
import numpy as np
```

### Paso 1. carga, limpieza y trasnformación de datos

```python
df_canada = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship',
                         skiprows = range(20),
                         skipfooter =2)
df_canada
```

```python
print(df_canada.shape)
```

```python
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_canada.head(5)
```

```python
df_canada.rename(columns = {'OdName':'Country', 'AreaName':'Continent',
                            'RegName':'Region'}, inplace = True)
df_canada.head(5)
```

```python
# Verificar que todas la columnas sena de tipo String
all(isinstance(column,str) for column in df_canada.columns)
```

```python
# Cambiar todas las columnas a String
df_canada.columns = list(map(str, df_canada.columns))
```

```python
all(isinstance(column,str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace = True)
df_canada.head(5)
```

```python
df_canada['Total'] = df_canada.sum(axis=1)
df_canada.head(5)
```

```python
years = list(map(str, range(1980, 2014)))
years
```

### Paso 2. Visualizacion de datos con matplotlib

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

#### Graficoso de Area.

```python
df_canada.sort_values(['Total'],ascending=False, axis=0, inplace=True)
df_canada.head(10)
```

```python
df_top5 = df_canada.head(5)
df_top5
```

```python
#Trasnponer renglones y columnas
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.index = df_top5.index.map(int)
df_top5
```

```python
df_top5.plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Tendencias de inmigracion de los primeros 5 paises a Canada')
plt.ylabel('NUmero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
# Transparencia
df_top5.plot(kind='area', alpha=0.25,stacked=False,figsize=(20,10))
plt.title('Tendencias de inmigracion de los primeros 5 paises a Canada')
plt.ylabel('NUmero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
ax = df_top5.plot(kind = 'area', alpha=0.35,figsize=(20,10))
ax.set_title('Tendencias de inmigracion de los primeros 5 paises a Canada')
ax.set_ylabel('NUmero de inmigrantes')
ax.set_xlabel('Años')
ax.show()
```

### Graficos de barras

```python
df_chile = df_canada.loc['Chile',years]
df_chile.head(5)
```

```python
# En los diagramas de barras el eje X se usa apra etiquetar y el eje Y
# se usa para la longitud de la barra
df_chile.plot(kind='barh',figsize=(10,10))
plt.xlabel('Numero de inmigrantes')
plt.ylabel('Años')
plt.title('Inmigrantes Chilenos a Canada desde 1980 a 2013')
for index, value in enumerate (df_chile):
    label = format(int(value), ',')
    plt.annotate(label, xy=(value, index), color='black')

plt.show()
```

```python
df_mexico = df_canada.loc['Mexico',years]
df_mexico.head(5)
```

```python
# En los diagramas de barras el eje X se usa apra etiquetar y el eje Y
# se usa para la longitud de la barra
df_mexico.plot(kind='bar',figsize=(10,10))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrantes Mexicanos a Canada desde 1980 a 2013')
plt.show()
```

### Graficos Circulares

```python
df_continentes = df_canada.groupby('Continent', axis=0).sum()
```

```python
df_continentes['Total'].plot(kind='pie',figsize =(5,6),autopct='%1.1f%%',
                             startangle=0,
                             shadow=True)
plt.title('Inmigracion a Canada por continente 1980-2013')
plt.show()
```

```python
explode_list=[0.1, 0, 0, 0, 0.1,0.1]

df_continentes['Total'].plot(kind='pie',figsize =(15,6),autopct='%1.1f%%',
                             startangle=0,
                             labels=None,
                             pctdistance = 1.12,
                             explode = explode_list)
plt.title('Inmigracion a Canada por continente 1980-2013')
plt.axis('equal')
plt.legend(labels= df_continentes.index, loc='lower left')
plt.show()
```
